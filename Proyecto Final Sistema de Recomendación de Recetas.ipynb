```python
#FUNCION NUTRIENTES

def scrape_nutriente(url):    
    request = requests.get(url) 
    soup_2 = BeautifulSoup(request.text,'html.parser')
    nutritions_section = soup_2.find_all("div", attrs={"id": 'nutritions-desktop'})[0]


# Encontrar cada elemento de información nutricional
    calories = nutritions_section.find('dt', text='Calorías').find_next_sibling('dd').text.strip()
    calories = calories[20:].strip()
    protein = nutritions_section.find('dt', text='Proteína').find_next_sibling('dd').text.strip()
    carbs = nutritions_section.find('dt', text='Carbohidratos').find_next_sibling('dd').text.strip()
    fat = nutritions_section.find('dt', text='Grasa').find_next_sibling('dd').text.strip()
    fiber = nutritions_section.find('dt', text='Fibra').find_next_sibling('dd').text.strip()
    #sodio = nutritions_section.find('dt', text='Sodio').find_next_sibling('dd').text.strip()
    #grasa_saturada = nutritions_section.find('dt', text='Grasa saturada').find_next_sibling('dd').text.strip()


        
    return calories, protein, carbs, fat, fiber
```

```python
#FUNCION DE RECOMENDACION BASE RECETAS

def recipe_recommend(df, number_of_dishes, number_of_candidates, nut_conf):
    tmp = df_p2.copy()
    candidates_list = []

    # create recommend-list {number_of_candidates} times
    for i in range(0,number_of_candidates):
        m = pl.LpProblem(sense=pl.LpMaximize) 
        tmp.loc[:,'v'] = [pl.LpVariable('x%d'%i, cat=pl.LpBinary) for i in range(len(tmp))]

        # limit the number of recipes
        m += pl.lpSum(tmp["v"]) <= number_of_dishes

        # limit the calories
        m += pl.lpDot(tmp["calories"], tmp["v"]) >= nut_conf["cal_lo"]
        m += pl.lpDot(tmp["calories"], tmp["v"]) <= nut_conf["cal_up"]
        #m += pl.value(tmp["Ingredientes"] == 'pollo')

        # limit the protein
        m += pl.lpDot(tmp["protein"], tmp["v"]) >= nut_conf["pro_lo"]
        m += pl.lpDot(tmp["protein"], tmp["v"]) <= nut_conf["pro_up"]

        # limit the fat (sum of fat / calories)
        m += pl.lpDot(tmp["fat"], tmp["v"]) >= nut_conf["fat_lo"]
        m += pl.lpDot(tmp["fat"], tmp["v"]) <= nut_conf["fat_up"]


        m.solve(pl.PULP_CBC_CMD(msg=0, options=['maxsol 1']))
        
        if m.status == 1:
            tmp.loc[:,'val'] = tmp["v"].apply(lambda x: pl.value(x))
            ret = tmp.query('val==1')["title"].values
            candidates_list.append(ret)
            # update dataframe (remove recommended title )
            tmp = tmp.query('val==0')
            
    return candidates_list
```
